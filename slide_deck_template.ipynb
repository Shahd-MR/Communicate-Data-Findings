## Investigation Overview

This presentation will help answer the following questions about the Blue Bike system:
- The most popular months for blue bike rides.
- What is the busiest day of the week for blue bike rides?
- What is the most frequent user type?
- What is the average duration of blue bike trips?
- Bike riding patterns for subscribers and non-subscribers.

## Dataset Overview

Blue Bike is a puplic service provided in Boston state, where poeple rent bikes for a single trip, or buy a monthly/annualy subscribtion, with more than 300 stations across Boston, Brookline, Cambridge, Everett, and Somerville. This dataset contains records of Blue Bike trips from January to May of the year 2019.

```python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline

# suppress warnings from final output
import warnings
warnings.simplefilter("ignore")
```

```python
# load in the dataset into a pandas dataframe
df_clean = pd.read_csv('blue_bike_clean')

# Load in output_toggle.tpl file
output_toggle = pd.read_csv('output_toggle.tpl', encoding = 'utf-8')
```

## Distribution of Blue Bike Trips Over the Months

The busiest month for bike rides is May with rides more than 200k, then comes April with more than 150k rides. January has the least bike rides, which is approximatly 70k. This shows that people prefer to take bike trips when the weather is warm. 

```python
plt.figure(figsize=[8, 5])
default_color = sb.color_palette()[0]
sb.countplot(data = df_clean, x = 'month', color = default_color)
plt.xticks([0, 1, 2, 3, 4], ['January', 'February', 'March', 'April', 'May'])
plt.title('Distribution of Blue Bike Trips Over the Months')
plt.xlabel('Month')
plt.ylabel('Count of Bike Trips')
plt.show()
```

## Distribution of Blue Bike Trips Over Weekdays

Interestingly, the distribution of bike rides over the days of the week is nearly the same. Except for saturday and sunday, with sunday having the least number of rides.

```python
plt.figure(figsize=[10, 5])
default_color = sb.color_palette()[0]
sb.countplot(data = df_clean, x = 'day_of_week', color = default_color)
plt.xticks([0,1,2,3,4,5,6],['Mon','Tue','Wed','Thu','Fri','Sat','Sun'])
plt.title('Distribution of Blue Bike Trips Over Weekdays')
plt.xlabel('Day of Week')
plt.ylabel('Count of Bike Trips')
plt.show()
```

## User Type Distribution
The user type bar plot shows that subscribers are approximatly five times more than the single trip riders (i.e. Costumers).

```python
plt.figure(figsize=([8,5]))
plt.title('User Type Distribution')
sb.countplot(data = df_clean, x = 'usertype', color = default_color)
plt.xlabel('User Type')
plt.ylabel('Count of Users')
plt.show()
```

## Average Duration per Day of Week
The week ends , Saturday and Sunday has the highest trip duration averages of approximatly 21 min,and 20 min respectively. The work days of the week show a nearly similar average.

```python
base_color = sb.color_palette()[0]
plt.figure(figsize=([8,5]))
sb.pointplot(data=df_clean, x= 'day_of_week', y='tripduration_min', color=base_color, ci='sd', linestyles="")
plt.xticks([0,1,2,3,4,5,6],['Mon','Tue','Wed','Thu','Fri','Sat','Sun'])
plt.xticks(rotation=15)
plt.title('Average Duration per Day of Week')
plt.xlabel('Day of Week')
plt.ylabel('Avg. Trip Duration (minutes)');
```

## Average Duration per User Type
Interestingly, the average duration of trips for single ride customers is much more higher than that for the subsicribers, with average durations of ~30 minutes and ~13 minutes respectivly.

```python
base_color = sb.color_palette()[0]
plt.figure(figsize=([8,5]))
sb.pointplot(data=df_clean, x= 'usertype', y='tripduration_min', color=base_color, ci='sd', linestyles="")
plt.xticks(rotation=15)
plt.title('Average Duration per User Type')
plt.xlabel('User Type')
plt.ylabel('Avg. Trip Duration (minutes)');
```

## User Type Distribution Over Weekdays
It seems that single ride customers prefer taking the bike trip during the weekends, Saturday and Sunday. During weekends, customer bike trips reached almost 30k, while on workdays it is around 10k. Subscribers on the other hand, seem to prefer workdays over weekends for the bike trips. Wednesday and Thursday are the most popular days for subsicribers, reaching more than 90k trips.

```python
plt.figure(figsize=[8, 5])
sb.countplot(data = df_clean, x = 'day_of_week', hue = 'usertype').legend(title = 'User Type')
plt.xticks([0,1,2,3,4,5,6],['Mon','Tue','Wed','Thu','Fri','Sat','Sun'])
plt.title('User Type Distribution Over Weekdays')
plt.xlabel('Day of Week')
plt.ylabel('Count of Users');
```

## Average Duration per User Type Over Weekdays
On average, subscribers take a slightly longer trips in weekends over workdays. Customers alos take logner trips during the weekends. The average duration for subscribers is idintical over the workdays, while for costomers Monday has the highest duration.

```python
plt.figure(figsize=([8,5]))
ax = sb.barplot(data = df_clean, x = 'day_of_week', y = 'tripduration_min', hue = 'usertype')
ax.legend(loc = 8, ncol = 3, framealpha = 1, title = 'User Type')
plt.xticks([0,1,2,3,4,5,6],['Mon','Tue','Wed','Thu','Fri','Sat','Sun'])
plt.title('Average Duration per User Type Over Weekdays')
plt.xlabel('Day of Week')
plt.ylabel('Avg. Trip Duration (minutes)');
```
